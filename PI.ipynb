```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
import numpy as np
```

# Lendo o CSV

```python
df = pd.read_csv("resultado.csv", sep=";", encoding="latin1")
```

```python
# Informações do dataframe
df.info()
```

```python
df.tail()
```

```python

# Encontrar todas as linhas que atendem às condições especificadas
df.loc[(df[' Município'] == 'VIANOPOLIS') & (df[' Conteudo'] != 'Valor aprovado')]
```

```python
# Adicionar nova coluna 'Valor Aprovado' com a condição especificada
df['Valor Aprovado'] = df[' Total'].where(df[' Conteudo'] == 'Quantidade aprovada', None)

df.tail()
```

```python
# Adicionar nova coluna 'Valor Aprovado' com a condição especificada
df['Quantidade Aprovada'] = df[' Total'].where(df[' Conteudo'] == 'Valor aprovado', None)

df.tail()
```

```python
nan_values = df.isna()
print(nan_values)
```

```python
any_missing_values = nan_values.any().any()

any_missing_values_in_column = nan_values.any()

missing_value_count = df.isnull().sum()

for column, has_missing in any_missing_values_in_column.items():
    if has_missing:
        count = missing_value_count[column]
        print(f"-----> Coluna '{column}' tem {count} missing values.")
        
print("\nMissing Values em todo o dataframe?")
print(any_missing_values)

print("\nMissing Values em cada coluna?")
print(any_missing_values_in_column)

print("\nMissing Value Counts em cada coluna:")
print(missing_value_count)
```

```python
# Dimensões do dataframe
df.shape
```

```python
# Transformando "-" para nan
df = df.replace(0, np.nan)
df.head()
```

```python
df['Código Município'] = df['Código Município'].astype(str)
df[' Ano'] = df[' Ano'].astype(str)
```

```python
# Gráfico de dispersão entre 'Diagnóstico por radiologia' e 'Total'
plt.figure(figsize=(10, 5))
plt.scatter(df[' Diagnóstico por radiologia'], df[' Total'], alpha=0.5)
plt.title('Dispersão: Diagnóstico por Radiologia vs Total')
plt.xlabel('Diagnóstico por Radiologia')
plt.ylabel('Total')
plt.show()
```

```python
# Filtrar valores nulos para as colunas de interesse
x = df[' Diagnóstico por radiologia'].dropna()
y = df[' Total'][x.index].dropna()

# Ajustar o tamanho das amostras
x = x.values.reshape(-1, 1)
y = y.values

# Ajustar o modelo de regressão linear
model = LinearRegression()
model.fit(x, y)

# Plotar a linha de regressão
plt.figure(figsize=(10, 5))
plt.scatter(x, y, color='blue', alpha=0.5)
plt.plot(x, model.predict(x), color='red', linewidth=2)
plt.title('Regressão Linear: Diagnóstico por Radiologia vs Total')
plt.xlabel('Diagnóstico por Radiologia')
plt.ylabel('Total')
plt.show()

# Coeficientes da regressão
model.coef_, model.intercept_
```

## Identificar as colunas relacionadas a procedimentos cirúrgicos

```python
# Listar colunas relacionadas a procedimentos cirúrgicos
surgical_cols = [col for col in df.columns if 'cirurgia' in col.lower()]
surgical_cols
```

```python
# Listar atributos categóricos
print("\nAtributos categóricos:")

# Listar colunas categóricas principais
categorical_cols = [' Município', ' Conteudo']

for col in categorical_cols:
    print(f"\nColuna: {col}")
    print("Valores únicos e suas frequências:")
    print(df[col].value_counts())
```

```python
# Remover espaços extras dos nomes das colunas
df.columns = df.columns.str.strip()

diagnostics_columns = [
    'Diagnóstico em laboratório clínico', 'Diagnóstico por radiologia',
    'Diagnóstico por tomografia',
    'Diagnóstico por ressonância magnética', 'Diagnóstico por medicina nuclear in vivo',
    'Diagnóstico por endoscopia', 'Diagnóstico por radiologia intervencionista',
    'Métodos diagnósticos em especialidades', 'Diagnóstico e procedimentos especiais em hemoterapia',
    'Diagnóstico por teste rápido'
]

# Configurar o layout para múltiplos gráficos em um único gráfico composto
fig, axes = plt.subplots(nrows=6, ncols=4, figsize=(20, 30))
fig.subplots_adjust(hspace=0.5, wspace=0.5)

# Iterar sobre as colunas de procedimentos e os eixos para plotar os histogramas
for ax, column in zip(axes.flatten(), diagnostics_columns):
    df[column].dropna().hist(bins=30, ax=ax)
    ax.set_title(column)
    ax.set_xlabel('Valores')
    ax.set_ylabel('Frequência')
    ax.grid(False)

plt.show()

# Criar boxplots para cada coluna de procedimentos em um único gráfico composto
plt.figure(figsize=(20, 10))
df[diagnostics_columns].boxplot()
plt.title('Boxplots de Diagnostico')
plt.xticks(rotation=90)
plt.ylabel('Valores')
plt.show()
```

**Desigualdade na Distribuição de Serviços:**
Existe uma desigualdade significativa na distribuição dos diagnósticos entre os municípios. Enquanto a maioria possui números baixos, alguns municípios têm valores muito altos, indicando possível concentração de serviços em centros urbanos maiores.


**Acessibilidade e Utilização:**
Alguns diagnósticos, como por ressonância magnética e medicina nuclear in vivo, são menos acessíveis ou menos utilizados, com muitos municípios sem registros desses serviços.


**Necessidade de Foco em Outliers:**
Os outliers indicam municípios com alta demanda ou melhor acesso a certos serviços de diagnóstico. Investigar esses municípios pode fornecer insights sobre boas práticas ou áreas com maior necessidade.
