```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Lendo o CSV

```python
df = pd.read_csv("resultado.csv", sep=";", encoding="latin1")
```

```python
# Informações do dataframe
df.info()
```

```python
df.columns
```

```python
nan_values = df.isna()
print(nan_values)
```

```python
any_missing_values = nan_values.any().any()

any_missing_values_in_column = nan_values.any()

missing_value_count = df.isnull().sum()

for column, has_missing in any_missing_values_in_column.items():
    if has_missing:
        count = missing_value_count[column]
        print(f"-----> Coluna '{column}' tem {count} missing values.")
        
print("\nMissing Values em todo o dataframe?")
print(any_missing_values)

print("\nMissing Values em cada coluna?")
print(any_missing_values_in_column)

print("\nMissing Value Counts em cada coluna:")
print(missing_value_count)
```
